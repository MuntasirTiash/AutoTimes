```python
from data_provider.data_loader import Dataset_EPSPanel

root_path = "/largessd/home/muntasir/Desktop/AutoTimes/dataset/panel/univariate"   
data_path = "df_10.csv"

seq_len, label_len, pred_len = 32, 4, 4

train_ds = Dataset_EPSPanel(
    root_path=root_path,
    flag='train',
    size=(seq_len, label_len, pred_len),
    data_path=data_path,
    train_end='2014-12-31',
    val_end='2017-12-31',
)

print("Train samples:", len(train_ds))

# Inspect one sample
x, y, x_mark, y_mark, firm_id, target_quarters = train_ds[0]
print("x shape:", x.shape)              # (seq_len, 1)
print("y shape:", y.shape)              # (label_len + pred_len, 1)
print("firm_id:", firm_id)
print("target_quarters:", target_quarters)
```

```python
import numpy as np

# collect some y values from train_ds in scaled space
ys_scaled = []
for i in range(min(200, len(train_ds))):
    _, y_scaled, _, _, _, _ = train_ds[i]
    ys_scaled.append(y_scaled[:, 0])   # flatten

ys_scaled = np.concatenate(ys_scaled)

print("Approx mean(z):", ys_scaled.mean())
print("Approx std(z):", ys_scaled.std())
print("Cached eps_mean:", train_ds.eps_mean)
print("Cached eps_std :", train_ds.eps_std)
```

```python
import random
idx = random.randint(0, len(train_ds) - 1)

x_scaled, y_scaled, _, _, firm_id, target_quarters = train_ds[idx]

y_scaled_vals = y_scaled[:, 0]
y_orig_via_inverse = train_ds.inverse_transform(y_scaled_vals)

print("Firm:", firm_id)
print("Target quarters:", target_quarters)
print("Scaled y (first 5):", y_scaled_vals[:5])
print("Inverse y (first 5):", y_orig_via_inverse[:5])
```

```python
import pandas as pd

df = pd.read_csv(f"{root_path}/{data_path}")
df[train_ds.date_col] = pd.to_datetime(df[train_ds.date_col])
df = df.sort_values([train_ds.firm_col, train_ds.date_col])

mask = (df[train_ds.firm_col] == firm_id) & \
       (df[train_ds.date_col].isin(target_quarters))
df_raw_segment = df.loc[mask].sort_values(train_ds.date_col)

print("Raw EPS from CSV:", df_raw_segment[train_ds.target_col].values)
print("Inverse y aligned :", y_orig_via_inverse)
```

```python
val_ds = Dataset_EPSPanel(
    root_path=root_path,
    flag='val',
    size=(seq_len, label_len, pred_len),
    data_path=data_path,
    train_end='2014-12-31',
    val_end='2017-12-31',
)

test_ds = Dataset_EPSPanel(
    root_path=root_path,
    flag='test',
    size=(seq_len, label_len, pred_len),
    data_path=data_path,
    train_end='2014-12-31',
    val_end='2017-12-31',
)

print("Val samples:", len(val_ds))
print("Test samples:", len(test_ds))

# Check that val/test target_quarters are in the right date ranges
_, _, _, _, _, tq_val = val_ds[0]
_, _, _, _, _, tq_test = test_ds[0]
print("Val target quarter range:", tq_val[0], "->", tq_val[-1])
print("Test target quarter range:", tq_test[0], "->", tq_test[-1])
```
